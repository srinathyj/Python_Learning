```python
#Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.
def upper_lower(mixed_case):
    l=0
    u=0
    ll=[]
    ul=[]
    for s in mixed_case:        
        if s.isupper():
            u += 1
            ul.append(s)
    
        elif s.islower():
            l += 1
            ll.append(s)
    
    print('{} Count of Upper Letters is :{}'.format(ul,u))
    print('{} Count of Lower Letters is :{}'.format(ll,l))
    
upper_lower("This")
```

```python
#Write a Python function that takes a list and returns a new list with unique elements of the first list.
def rem_dup(dup_list):
    dedup_set=set(dup_list)
    return list(dedup_set)

result=rem_dup([1,2,3,4,4,2,1,3])
print(result)
    
```

```python
#Write a Python function that takes a number as a parameter and check the number is prime or not.
def is_prime(my_number):
    for n in my_number:
        if n > 1:
            for i in range(2,n):
                if (n % i) == 0:
                    print(n, "is not a prime number")
                    break
            else:
                print(n, "is a prime number")
        else:
            print(n, "is not a prime number")

is_prime([1,2,2,3,4,5])
```

```python
#Write a Python program to print the even numbers from a given list.

def is_even(my_even_list):
    for n in my_even_list:
        if n % 2 == 0:
            print(n,': is an Even number')
        else:
            print(n,': is an Odd number')
            
is_even([2,3,4,5,6,7,8])
```

```python
#Write a Python function that checks whether a passed string is palindrome or not.

def is_palindrome(my_pal_str):
    if my_pal_str[0::].lower() == my_pal_str[::-1].lower():
        print(my_pal_str,' Is a Paldrome')
    else:
        print(my_pal_str,' Is not a Paldrome')

is_palindrome("Malayalam")
is_palindrome("Kayak")
is_palindrome("Srinath")
is_palindrome("Chandra")
```
